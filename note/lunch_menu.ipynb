```python
start_idx = df.columns.get_loc('2025-01-07')

for _, row in df.iterrows():  # 모든 행을 순회
    for c in df.columns[start_idx:]:  # 지정한 컬럼부터 마지막까지 순회
        print(f"{row['ename']},{c},{row[c]}")
```

```python
melted_df = df.melt(id_vars=['ename'], value_vars=df.columns[start_idx:-2], 
                     var_name='dt', value_name='menu')  #https://pandas.pydata.org/docs/reference/api/pandas.melt.html
```

```python
for i in melted_df.index:  # 행 인덱스를 직접 사용
    ename = melted_df.loc[i, "ename"]
    dt = melted_df.loc[i, "dt"]
    value = melted_df.loc[i, "menu"]
    print(f"{ename},{dt},{value}")
```

```python
for i in melted_df.index:  # 정수 인덱스를 사용해서 ...
    ename = melted_df.iloc[i, 0]
    dt = melted_df.iloc[i, 1]
    value = melted_df.iloc[i, 2]
    print(f"{ename},{dt},{value}")
```

```python
for i in range(len(melted_df)):  # 정수 인덱스를 사용해서 ...
    ename = melted_df.iat[i, 0]
    dt = melted_df.iat[i, 1]
    value = melted_df.iat[i, 2]
    print(f"{ename},{dt},{value}")
```
